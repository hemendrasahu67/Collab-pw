```python
# @title **Json Links Extractor Script**

import json
from google.colab import files

dict = {}

user = str(input("Enter Json Here : "))
site = json.loads(user)
for data in site["data"]:

  cstr = data["tags"][0]["name"]
  chap = cstr.replace('+',' ')
  chap = chap.replace('-',' ')
  chap = chap.replace('_',' ')
  chap = chap.replace('*',' ')
  chap = chap.replace('!',' ')
  chap = chap.replace('@',' ')
  chap = chap.replace('#',' ')
  chap = chap.replace('$',' ')
  chap = chap.replace('%',' ')
  chap = chap.replace('^',' ')
  chap = chap.replace('&',' ')
  chap = chap.replace(':',' ')
  chap = chap.replace('<',' ')
  chap = chap.replace('>',' ')
  chap = chap.replace('?',' ')
  chap = chap.replace(';',' ')
  chap = chap.replace(',',' ')
  chap = chap.replace('.',' ')
  chap = chap.replace('(',' ')
  chap = chap.replace(')',' ')
  chap = chap.replace('[',' ')
  chap = chap.replace(']',' ')
  chap = chap.replace('{',' ')
  chap = chap.replace('}',' ')
  chap = chap.replace('=',' ')
  chap = chap.replace('|',' ')
  chap = ' '.join(chap.split())

  nstr = data["topic"]
  name = nstr.replace('+',' ')
  name = name.replace('-',' ')
  name = name.replace('_',' ')
  name = name.replace('*',' ')
  name = name.replace('!',' ')
  name = name.replace('@',' ')
  name = name.replace('#',' ')
  name = name.replace('$',' ')
  name = name.replace('%',' ')
  name = name.replace('^',' ')
  name = name.replace('&',' ')
  name = name.replace(':',' ')
  name = name.replace('<',' ')
  name = name.replace('>',' ')
  name = name.replace('?',' ')
  name = name.replace(';',' ')
  name = name.replace(',',' ')
  name = name.replace('.',' ')
  name = name.replace('(',' ')
  name = name.replace(')',' ')
  name = name.replace('[',' ')
  name = name.replace(']',' ')
  name = name.replace('{',' ')
  name = name.replace('}',' ')
  name = name.replace('=',' ')
  name = name.replace('|',' ')
  name = ' '.join(name.split())

  code = data["url"][38:74:1]
  link = f'https://pwjarvis.onrender.com?v=https://d1d34p8vz63oiq.cloudfront.net/{code}/master.mpd&quality=720"'
  dict[name] = link

fill = ''
fill += f'@echo off\n'
fill += f'setlocal enabledelayedexpansion\n'
fill += f'\nset "chapter={chap}"\n\n'
fill += f'mkdir "!chapter!" >nul 2>nul\n'
fill += f'cd "!chapter!"\n'
fill += f'mkdir "Lecture" >nul 2>nul\n'

for i, (keys, values) in enumerate(reversed(dict.items()),start=1):
  fill += f'\nset "name{i}={keys}"\n'
  fill += f'set "vurl{i}={values}\n'
  fill += f'if not exist "Lecture\!name{i}!.mp4" (\n'
  fill += f'    rmdir /s /q "!name{i}!" >nul 2>nul\n'
  fill += f'    del master.m3u8 >nul 2>nul\n'
  fill += f'    title Download Initiated.. !name{i}!.mp4\n'
  fill += f'    echo.\n'
  fill += f'    echo  Processing.. {chap} - {keys}\n'
  fill += f'    echo.\n'
  fill += f'    curl -X GET "!vurl{i}!" -o "master.m3u8"\n'
  if i < 10:
    fill += f'    N_m3u8DL-RE.exe "master.m3u8" --save-dir "Lecture" --save-name "L0{i}"\n'
  else:
    fill += f'    N_m3u8DL-RE.exe "master.m3u8" --save-dir "Lecture" --save-name "L{i}"\n'
  fill += f'    del master.m3u8\n'
  fill += f'    cls\n'
  fill += f') else (\n'
  fill += f'    echo Lecture exists, !name{i}! skipping downloading.\n'
  fill += f')\n'

with open(f'{chap}.bat','w') as f:
  f.write(fill)
files.download(f'/content/{chap}.bat')
```
