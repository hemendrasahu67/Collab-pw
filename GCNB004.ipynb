```python
import json
from google.colab import files

dict1 = {}
dict2 = {}
dict3 = {}
dict4 = {}

Json1 = json.loads(str(input("Enter VLEC Json Here : ")))
Json2 = json.loads(str(input("Enter VDPP Json Here : ")))
Json3 = json.loads(str(input("Enter NLEC Json Here : ")))
Json4 = json.loads(str(input("Enter NDPP Json Here : ")))

for items in Json1["data"]:

  cstr = items["tags"][0]["name"]
  chap = cstr.replace('_',' ')
  chap = chap.replace(':',' ')
  chap = chap.replace('-',' ')
  chap = chap.replace('(',' ')
  chap = chap.replace(')',' ')
  chap = chap.replace('|',' ')
  chap = ' '.join(chap.split())

  nstr = items["topic"]
  nlec = nstr.replace('_',' ')
  nlec = nlec.replace(':',' ')
  nlec = nlec.replace('-',' ')
  nlec = nlec.replace('(',' ')
  nlec = nlec.replace(')',' ')
  nlec = nlec.replace('|',' ')
  nlec = ' '.join(nlec.split())
  dict1[nlec] = f'https://api.pwjarvis.tech?v={items["url"][38:74:1]}&quality=720'

for items in Json2["data"]:

  nstr = items["topic"]
  ndpp = nstr.replace('_',' ')
  ndpp = ndpp.replace(':',' ')
  ndpp = ndpp.replace('-',' ')
  ndpp = ndpp.replace('(',' ')
  ndpp = ndpp.replace(')',' ')
  ndpp = ndpp.replace('|',' ')
  ndpp = ' '.join(ndpp.split())
  dict2[ndpp] = f'https://api.pwjarvis.tech?v={items["url"][38:74:1]}&quality=720'

for items in Json3["data"]:

  nstr = items['homeworkIds'][0]['attachmentIds'][0]['name']
  burl = items['homeworkIds'][0]['attachmentIds'][0]['baseUrl']
  bkey = items['homeworkIds'][0]['attachmentIds'][0]['key']

  slec = nstr.replace(' ','')
  slec = slec.replace(':','_')
  slec = slec.replace('-','_')
  slec = slec.replace('(','_')
  slec = slec.replace(')','_')
  slec = slec.replace('|','_')
  slec = ' '.join(slec.split())

  dict3[slec] = f'{burl}{bkey}'

for items in Json4["data"]:

  nstr = items['homeworkIds'][0]['attachmentIds'][0]['name']
  burl = items['homeworkIds'][0]['attachmentIds'][0]['baseUrl']
  bkey = items['homeworkIds'][0]['attachmentIds'][0]['key']

  sdpp = nstr.replace(' ','')
  sdpp = sdpp.replace(':','_')
  sdpp = sdpp.replace('-','_')
  sdpp = sdpp.replace('(','_')
  sdpp = sdpp.replace(')','_')
  sdpp = sdpp.replace('|','_')
  sdpp = ' '.join(sdpp.split())

  dict4[sdpp] = f'{burl}{bkey}'

fill = ''
fill += f'@echo off\n'
fill += f'setlocal enabledelayedexpansion\n'
fill += f'\nset "chapter={chap}"\n\n'
fill += f'mkdir "!chapter!" >nul 2>nul\n'
fill += f'cd "!chapter!"\n'
fill += f'mkdir "L" >nul 2>nul\n'
fill += f'mkdir "D" >nul 2>nul\n'
fill += f'mkdir "N" >nul 2>nul\n'
fill += f'mkdir "S" >nul 2>nul\n'

for i, (keys, values) in enumerate(reversed(dict1.items()),start=1):
  fill += f'\nset "name{i}={keys}"\n'
  fill += f'set "lurl{i}={values}"\n'
  fill += f'if not exist "L\!name{i}!.mp4" (\n'
  fill += f'    rmdir /s /q "!name{i}!" >nul 2>nul\n'
  fill += f'    del master.m3u8 >nul 2>nul\n'
  fill += f'    title Download Initiated.. !name{i}!.mp4\n'
  fill += f'    echo.\n'
  fill += f'    echo  Processing.. {chap} - {keys}\n'
  fill += f'    echo.\n'
  fill += f'    curl -X GET "!lurl{i}!" -o "master.m3u8"\n'
  if i < 10:
    fill += f'    N_m3u8DL-RE.exe "master.m3u8" --save-dir "L" --save-name "L0{i}"\n'
  else:
    fill += f'    N_m3u8DL-RE.exe "master.m3u8" --save-dir "L" --save-name "L{i}"\n'
  fill += f'    del master.m3u8\n'
  fill += f'    cls\n'
  fill += f') else (\n'
  fill += f'    echo Lecture exists, !name{i}! skipping downloading.\n'
  fill += f')\n'

for i, (keys, values) in enumerate(reversed(dict2.items()),start=1):
  fill += f'\nset "name{i}={keys}"\n'
  fill += f'set "durl{i}={values}"\n'
  fill += f'if not exist "D\!name{i}!.mp4" (\n'
  fill += f'    rmdir /s /q "!name{i}!" >nul 2>nul\n'
  fill += f'    del master.m3u8 >nul 2>nul\n'
  fill += f'    title Download Initiated.. !name{i}!.mp4\n'
  fill += f'    echo.\n'
  fill += f'    echo  Processing.. {chap} - {keys}\n'
  fill += f'    echo.\n'
  fill += f'    curl -X GET "!durl{i}!" -o "master.m3u8"\n'
  if i < 10:
    fill += f'    N_m3u8DL-RE.exe "master.m3u8" --save-dir "D" --save-name "D0{i}"\n'
  else:
    fill += f'    N_m3u8DL-RE.exe "master.m3u8" --save-dir "D" --save-name "D{i}"\n'
  fill += f'    del master.m3u8\n'
  fill += f'    cls\n'
  fill += f') else (\n'
  fill += f'    echo Lecture exists, !name{i}! skipping downloading.\n'
  fill += f')\n'

for i, (keys, values) in enumerate(reversed(dict3.items()),start=1):
  fill += f'\nset "name{i}={keys}"\n'
  fill += f'set "nurl{i}={values}"\n'
  fill += f'if not exist "N\!name{i}!.mp4" (\n'
  fill += f'    rmdir /s /q "!name{i}!" >nul 2>nul\n'
  fill += f'    del master.m3u8 >nul 2>nul\n'
  fill += f'    title Download Initiated.. !name{i}!.mp4\n'
  fill += f'    echo.\n'
  fill += f'    echo  Processing.. {chap} - {keys}\n'
  fill += f'    echo.\n'
  fill += f'    curl -X GET "!nurl{i}!" -o "master.m3u8"\n'
  fill += f'    N_m3u8DL-RE.exe "master.m3u8" --save-dir "N" --save-name "{keys}"\n'
  fill += f'    del master.m3u8\n'
  fill += f'    cls\n'
  fill += f') else (\n'
  fill += f'    echo Lecture exists, !name{i}! skipping downloading.\n'
  fill += f')\n'

for i, (keys, values) in enumerate(reversed(dict4.items()),start=1):
  fill += f'\nset "name{i}={keys}"\n'
  fill += f'set "surl{i}={values}"\n'
  fill += f'if not exist "S\!name{i}!.mp4" (\n'
  fill += f'    rmdir /s /q "!name{i}!" >nul 2>nul\n'
  fill += f'    del master.m3u8 >nul 2>nul\n'
  fill += f'    title Download Initiated.. !name{i}!.mp4\n'
  fill += f'    echo.\n'
  fill += f'    echo  Processing.. {chap} - {keys}\n'
  fill += f'    echo.\n'
  fill += f'    curl -X GET "!surl{i}!" -o "master.m3u8"\n'
  fill += f'    N_m3u8DL-RE.exe "master.m3u8" --save-dir "S" --save-name "{keys}"\n'
  fill += f'    del master.m3u8\n'
  fill += f'    cls\n'
  fill += f') else (\n'
  fill += f'    echo Lecture exists, !name{i}! skipping downloading.\n'
  fill += f')\n'

with open(f'{chap}.bat','w') as f:
  f.write(fill)
files.download(f'/content/{chap}.bat')
```
